```python
import numpy as np
import pandas as pd
```

```python
filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/automobileEDA.csv"
```

```python
# Load the data and store it in dataframe df
df = pd.read_csv(filename, header=0)
```

```python
df.head()
```

```python
# Import visualization packages "Matplotlib" and "Seaborn"
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
# List the data types for each column
print(df.dtypes)
```

```python
# Data type of the "peak-rpm" column
df["peak-rpm"].dtypes
```

```python
df.corr()
```

```python
# Correlation between bore, stroke, compression-ratio and horsepower
df[["bore","stroke","compression-ratio","horsepower"]].corr()
```

```python
# Use a scatterplot to find relationship between "price" and "engine-size"
sns.regplot(x="engine-size", y="price", data=df)
plt.ylim(0,)
```

```python
# As the engine-size goes up, the price goes up: this indicates a positive direct correlation between these two variables. Engine size seems like a pretty good predictor of price since the regression line is almost a perfect diagonal line.
```

```python
df[["engine-size","price"]].corr()
```

```python
# Find out if Highway mpg is a predictor variable of price by generating a scatter-plot
sns.regplot(x="highway-mpg", y="price", data=df)
```

```python
# As highway-mpg goes up, the price goes down, indicating an inverse or negative relationship between these two variables.Highway mpg could be a predicator of price
```

```python
df[['highway-mpg', 'price']].corr()
```

```python
# Explore if "peak-rpm" is a predictor variable of "price"
sns.regplot(x="peak-rpm", y="price", data=df)
```

```python
# Peak rpm does not seem like a good predictor of the price at all since the regression line is close to horizontal. Also, the data points are very scattered and far from the fitted line, showing lots of variability. Therefore, it iss not a reliable variable.
```

```python
df[["peak-rpm", "price"]].corr()
```

```python
# Find out correlation between the variables "stroke" and "price"
df[["stroke", "price"]].corr()
```
