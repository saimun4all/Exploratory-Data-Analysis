# What are the main characteristics that have the most impact on the car price?

```python
import numpy as np
import pandas as pd
```

```python
filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/automobileEDA.csv"
```

```python
# Load the data and store it in dataframe df
df = pd.read_csv(filename, header=0)
```

```python
df.head()
```

```python
# Import visualization packages "Matplotlib" and "Seaborn"
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
# List the data types for each column
print(df.dtypes)
```

```python
# Data type of the "peak-rpm" column
df["peak-rpm"].dtypes
```

```python
df.corr()
```

```python
# Correlation between bore, stroke, compression-ratio and horsepower
df[["bore","stroke","compression-ratio","horsepower"]].corr()
```

```python
# Use a scatterplot to find relationship between "price" and "engine-size"
sns.regplot(x="engine-size", y="price", data=df)
plt.ylim(0,)
```

#### As the engine-size goes up, the price goes up: this indicates a positive direct correlation between these two variables. Engine size seems like a pretty good predictor of price since the regression line is almost a perfect diagonal line.


```python
df[["engine-size","price"]].corr()
```

```python
# Find out if Highway mpg is a predictor variable of price by generating a scatter-plot
sns.regplot(x="highway-mpg", y="price", data=df)
```

#### As highway-mpg goes up, the price goes down, indicating an inverse or negative relationship between these two variables.Highway mpg could be a predicator of price

```python
df[['highway-mpg', 'price']].corr()
```

```python
# Explore if "peak-rpm" is a predictor variable of "price"
sns.regplot(x="peak-rpm", y="price", data=df)
```

#### Peak rpm does not seem like a good predictor of the price at all since the regression line is close to horizontal. Also, the data points are very scattered and far from the fitted line, showing lots of variability. Therefore, it iss not a reliable variable.

```python
df[["peak-rpm", "price"]].corr()
```

```python
# Find out correlation between the variables "stroke" and "price"
df[["stroke", "price"]].corr()
```

```python
sns.regplot(x="stroke", y="price", data=df)
```

```python
# Find out relationship between Categorical variables and price
sns.boxplot(x="body-style", y="price", data=df)
```

#### We can see that the distribution of price between the different body-style categories have a significant overlap, so body-style would not be a good predicator of price. 

```python
# Find out relationship between "engine-location" and price
sns.boxplot(x="engine-location", y="price", data=df)
```

#### The distribution of price between these two engine-location categories, front and rear, are distinct enough to take engine-location as a potential good predictor of price.

```python
# Find out relationship between "drive-wheels" and price
sns.boxplot(x="drive-wheels", y="price", data=df)
```

#### The distribution of price between the different drive-wheels categories differs. As such, drive-wheels could potentially be a predictor of price.

## Grouping data

```python
df['drive-wheels'].unique()
```

```python
df_group_one = df[['drive-wheels', 'body-style', 'price']]
```

```python
# Grouping results
df_group_one = df_group_one.groupby(['drive-wheels'],as_index=False).mean()
df_group_one
```

```python
# The result shows that rear-wheel drive vehicles are, on average, the most expensive.
```

```python
# Grouping data based on "price" and "body-style"
df_group_two = df[['body-style','price']]
```

```python
df_group_two = df_group_two.groupby(['body-style'],as_index=False).mean()
df_group_two
```

```python
# grouping results
df_gptest = df[['drive-wheels','body-style','price']]
grouped_test1 = df_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean()
grouped_test1
```

```python
grouped_pivot = grouped_test1.pivot(index='drive-wheels',columns='body-style')
grouped_pivot
```

```python
grouped_pivot = grouped_pivot.fillna(0) #fill missing values with 0
grouped_pivot
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# Use heatmap to visualize the relationship between Body Style and Price
plt.pcolor(grouped_pivot, cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

#label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```
